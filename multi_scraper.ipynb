```python
news_title = []
source = []
typer = []
yf = requests.get("https://finance.yahoo.com/topic/stock-market-news/").text
soup = BeautifulSoup(yf, 'lxml')
for i in soup.find_all('h3', class_ = 'Mb(5px)'):
  news_title.append(i.text.strip())
for j in soup.find_all('div', class_ = 'Ov(h) Pend(44px) Pstart(25px)'):
  typer.append(j.text.strip())
for k in soup.find_all('div', class_ = 'C(#959595) Fz(11px) D(ib) Mb(6px)'):
  source.append(k.text.strip())
```

```python
len(news_title), len(typer), len(source)
```

```python
pd.DataFrame(data = {'Title': news_title, 'type': typer, 'source': source})
```
