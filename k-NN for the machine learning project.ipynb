```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("/content/TATAMOTORS.csv")
```

```python
X = data["Open"]
Y = data["Close"]
data["Profit"] = X - Y
data.info()
```

```python
class_profit = lambda x : 1 if x > 0 else 0
data["class_profit"] = data["Profit"].apply(class_profit)
```

```python
data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
features = ["Profit"]
scaled_values = scaler.fit_transform(data[features])
data["norm_Profit"] = scaled_values
```

```python
data.head()
```

```python
data.info()
# data.to_csv("C:\\D\ganja_op.csv")
```

```python
data["Profit_achieveable"] = data["High"] - data["Open"]
data["Loss_achieveable"] = data["Open"] - data["Low"]
data.info()
```

```python
data["Fight"] = data["High"] - data["Low"]
```

```python
data.info()
```

```python
np.corrcoef(data["Profit"], data["norm_Profit"])
```

```python
data_free = data
data_free = data_free.drop(columns = "Open")
data_free = data_free.drop(columns = "Close")
data_free = data_free.drop(columns = "High")
data_free = data_free.drop(columns = "Low")
data_free = data_free.drop(columns = "Adj Close")
```

```python
data_free.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_values = scaler.fit_transform(data_free[["Profit", "Profit_achieveable", "Loss_achieveable", "Fight"]])
data_free["norm_Profit"] = scaled_values[:, 0]
data_free["norm_Profit_achieaveable"] = scaled_values[:, 1]
data_free["norm_Loss_achieaveable"] = scaled_values[:, 2]
data_free["norm_Fight"] = scaled_values[:, 3]
```

```python
data_free.info()
data_free.to_csv("TATAMOTORS.csv")
```

```python
# splitting the data into test and trian data
from sklearn.model_selection import train_test_split
X = data_free[["norm_Profit_achieaveable"]]
Y = data_free['class_profit']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 100)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors = 3, metric = "euclidean")
model.fit(X_train, Y_train)
model.score(X_train, Y_train)
```
